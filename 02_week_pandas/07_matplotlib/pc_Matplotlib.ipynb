```python
#2.Loop over the years from 1880 to 2020
#3.Use your parse_dataset(year) function to parse each year
#4.Use the pandas.sum() method to calculate the sum of the frequencies (births)
#5.Append the result to the dictionary using birth_dict['total_births'].append()
#6.Also append the year to birth_dict['year']
#7.Create a yearly_births DataFrame by passing births_dictto pd.DataFrame()
#8.This should result in a DataFrame that has the year as one column and the 
#total_birthsin that year in another column.
```

```python
def parse_dataset(year):
    path = f"yob{year}.txt"
    df = pd.read_csv(path, names = ["name", "gender", "frequency"])
    sum_births = df["frequency"].sum()
    births_dict["total_births"].append(sum_births)
    births_dict["year"].append(year)
    
    return births_dict
```

```python
col_list=['name', 'gender', 'frequency']
```

```python
for year in range(1880, 2022):
    sum_births_per_year = parse_dataset(year)
```
