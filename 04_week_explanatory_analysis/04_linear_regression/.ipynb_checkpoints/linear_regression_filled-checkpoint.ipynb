## Ingredients for Linear Regression

### 1.1 Model Equation

**A LR model is a linear function that is fitted to the data**   
single feature/predictor ===>
$$ y = w_1 x + w_0 $$ 

multiple features/predictors ===>
$$ y = w_1 x_1 + w_2 x_2 + ... + w_n x_n + w_0 $$ 

### 1.2 Loss Function

**The loss function helps us evaluate how our LR model is doing**

$$ MSE = \frac{1}{N}\sum((y_{pred_i}-y_{true_i})^2) $$ 

Root Mean Square Error, is going to have the same units as my y

### Q: What is our target here ? What should the training process attempt to achieve in terms of MSE ?

### 1.3 Optimization Algorithm

**Our aim in training/fitting a LR model is to get the best model possible**
**The best model possible will have the lowest MSE possible for our data**
**This is done using one of these two options:**
    
- Analytical method: **Normal Equation**
            * Will get us the global minimum of MSE, i.e. best model possible
            * slow and expensive to calculate 
            * OLS
- Numerical: **Gradient Descent**
            * calculates loss function gradient, updates weights moving in direction of a better loss function
            * significantly faster than analytical approach
            * might not get us the best possible model, but will give us something close to it

## 2. Visual example using random datapoints
In this example we will start from a true relationship in the data and we will mimic possible set of observations. The line is representing what we normally don't have and want to predict using linear regression models.

```python
# Create random datapoints
x = np.random.uniform(low=0, high=100, size=200)
```

```python
x
```

```python
y = [1]*200
```

```python
w1 = 1
w0 = 2
noise = np.random.normal(loc=0, scale=30, size=200)
y = w1*x + w0 +noise
```

```python
plt.scatter(x,y)
plt.plot(x, 1*x+2, c='r')
```

### Create single feature relationship between x and y

$$ y = w_1 x + w_0 $$ 

## 3. Linear Regression implementation in statsmodels
Let's prepare a dataframe using previous dataset:


```python
# transform above data into a dataframe as statemodels expects a pandas dataframe as an input
```

```python
df = pd.DataFrame({'x':x, 'y':y})
```

```python
df.head()
```

```python
df.shape
```

#### We choose ols model from statsmodels, and provide formula and data for linear regression 

Note: OLS is just another name for Linear Regression

```python
model = sm.ols(formula='y ~ x', data=df).fit()  # instantiate and fit the model
```

#### Our coefficients and intercept could be extracted using the `params` property:

```python
model.params
```

```python
model.params['x']
```

```python
model.params['Intercept']
```

#### Let's visualize all results:

```python
plt.scatter(x,y)
plt.plot(x, w1*x+w0, c='r')
plt.plot(x, model.params['x']*x + model.params['Intercept'], c='g', label='model' )
```

## 4. Goodness of Fit

To examine how well a linear regression model is modeling our data we use R$^2$ (if there are more coefficients defined it's better to look at the Adjusted R$^2$ which is a variation of this metric). 
* It tells you how much variance in your data is explained by your model — high R$^2$ is good, because then most most of the variance in your data is explained by your model (maximum value is 1). 
* It tells you how much better your model is than predicting average $y$ — if your R$^2$ is 0 (minimum value) then your model is no better than predicting an average value of $y$ 

```python
model.rsquared
```

In `statsmodels` we could also print out the whole model summary that will provide us with a very detailed information about the model:

```python
model.summary()
```

## 5. Example on the Penguins Dataset

```python
penguins = sns.load_dataset('penguins')

penguins.head()
```

We can see above there are some observations that are missing data. Remove those!

```python
penguins.dropna(inplace=True)
```

#### Let's fit a LR model between the body mass of penguins and their body measurements

```python
## step 1: let's check the correlation between the body mass and the bill and flipper measurements

penguins.corr()
```

```python
## visualize the correlation with a heatmap

sns.heatmap(penguins.corr())
```

```python
# let's plot the relationships using sns.lmplot

sns.lmplot(data=penguins, x='bill_length_mm', y='body_mass_g', ci=None)
```

```python
sns.lmplot(data=penguins, x='bill_depth_mm', y='body_mass_g', ci=None)
```

```python
sns.lmplot(data=penguins, x='flipper_length_mm', y='body_mass_g', ci=None)
```

#### Building a linear regression model using one independent variable:

```python
model1 = sm.ols(formula='body_mass_g ~ bill_length_mm', data=penguins).fit()  # instantiate and fit the model
```

```python
model1.rsquared
```

#### How can you interpret the parameters (coefficient and intercept)?

#### Building a linear regression model using two independent variable:

```python
model2 = sm.ols(formula='body_mass_g ~ bill_depth_mm + bill_length_mm', data=penguins).fit()  
```

```python
model2.rsquared
```

#### How can you interpret the parameters (coefficient and intercept)?

#### Building a linear regression model using three independent variable:

```python
model3 = sm.ols(formula='body_mass_g ~ bill_depth_mm + bill_length_mm + flipper_length_mm', data=penguins).fit()  # instantiate and fit the model
```

```python
model3.rsquared
```

#### How can you interpret the parameters (coefficient and intercept)?

### How can we make this model even better ? What other penguin features might be useful ?

With linear regression we can also use categorical variables. In our case, we can use the `species` column.

```python
model4 = sm.ols(formula='body_mass_g ~ bill_depth_mm + bill_length_mm + flipper_length_mm + species', data=penguins).fit()  # instantiate and fit the model
```

```python
model4.rsquared
```
