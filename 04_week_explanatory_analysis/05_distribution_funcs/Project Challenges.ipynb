```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1.Filter bike data for one day

bike_data = pd.read_csv('../data/bikes_clean.csv', parse_dates=True, index_col=0)
```

```python
bike_data.head()
```

```python
bikes_data_20110506 = bike_data.loc['2011-05-06']
bikes_data_20110506.sample(10)
```

```python
#2.Using sns.displot() plot the distribution of the count during that day (Try setting the parameter kde=True). More info on this plotting function here: displot

sns.displot(bikes_data_20110506, x='count', kde = True, stat="probability");
```

```python
#3.Filter bike data for the year 2011
mask_2011 = bike_data['year'] == 2011
bike_data_2011 = bike_data[mask_2011]
bike_data_2011.sample()
```

```python
#4.Again using sns.displot() plot the distribution of the temperature of the entire year
```

```python
sns.displot(bike_data_2011, x='temp', kde = True, stat="probability");
```
