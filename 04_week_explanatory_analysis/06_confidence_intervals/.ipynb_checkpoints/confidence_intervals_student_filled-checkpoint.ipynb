## 1. Confidence Intervals calculation using Bootstrapping

### What does Confidence Interval (CI) mean ??

 - **Confidence Interval** is a **range of values** that is likely to include a population parameter with a certain **confidence level** (usually 95%)
 
 - In other words, confidence interval is simply a way to measure how well your sample represents the population you are studying

### How can we calculate the CI of the mean of a population from a sample ???

### Method 1: Bootstrap sampling

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# from seaborn import the penguin data

penguins = sns.load_dataset('penguins')
```

```python
penguins.info()
```

```python
len(penguins)
```

```python
# how many penguins species are in the dataset?

penguins['species'].unique()
```

```python
# create a dataframe with Adelie only

penguins_adelie = penguins[penguins['species'] == 'Adelie']
```

```python
# what are the min and max values of the bill length for Adelie?

penguins_adelie['bill_length_mm'].max(), penguins_adelie['bill_length_mm'].min()
```

### Example:  Example: Calculating the confidence interval of the mean of the bill length for the Adelie species

```python
# extract the data: bill_length_mm values

bill_length_adelie = penguins_adelie['bill_length_mm']


# calculate sample mean


bill_length_adelie.mean()
```

```python
# how many observations do we have? what is the size of our sample?

len(bill_length_adelie)
```

```python
# this also works

len(penguins_adelie)
```

```python
# how do we take a sample with replacement ??

# we actually want to get same size samples as the original sample, but with replacement. 

bootstrap_sample = bill_length_adelie.sample(frac=1, replace=True)

print(len(bootstrap_sample), bootstrap_sample.mean())
```

```python
# initialize sample means list

sample_means_list = []


## let's take 10_000 bootstrap samples and record their means

for i in range(10_000):
    bootstrap_sample = bill_length_adelie.sample(frac=1, replace=True)
    sample_means_list.append(bootstrap_sample.mean())
```

```python
sample_means_list
```

```python
# let's plot the histogram of the sample means
plt.figure(figsize=(12,4))
plt.hist(sample_means_list, bins=40);
```

```python
# let's plot the distplot of the sample means

sns.displot(sample_means_list, bins=40);
```

```python
# same can be plotted for probability

sns.displot(sample_means_list, bins=40, stat='probability', kde=True);

# KDE = "technique that lets you create a smooth curve given a set of data."
# https://mathisonian.github.io/kde/
```

```python
# let's now calculate the 95% confidence interval using pd.quantile()

alpha = 0.95

lower_quantile = (1-alpha)/2  

upper_qunatile = alpha + ((1-alpha)/2)

print(alpha, lower_quantile, upper_qunatile)
```

## Method 2. Confidence Intervals calculation using z-value formula

Another method to calculate the confidence interval of the population mean from a sample is to use the following formula:

$$  CI = \bar{x} \pm z_{c}\frac{\sigma}{\sqrt{n}} $$

where $ \bar{x} $ is the sample mean, $ z_{c} $ is the Z value for the confidence level we require, $ \sigma $ is the sample standard deviation and $ n $ is the size of our sample. The second term of the equation is also called the **standard error**.

This equation is build on the **Central Limit Theorem**, which states that states that the distribution of sample means approximates a **Normal Distribution** as the sample size $ n $ gets larger, regardless of distribution of the original population. Sample sizes equal to or greater than 30 are often considered sufficient for the CLT to hold.

### Example: let's recalculate the confidence interval of of the mean of the sepal length for the setosa species

```python
# sample mean and std 

# calculate mean of bill length for Adelie:
sample_mean = bill_length_adelie.mean()

# calculate std of bill length for Adelie:
sample_std = bill_length_adelie.std()

print(sample_mean, sample_std)
```

$$  CI = \bar{x} \pm z_{c}\frac{\sigma}{\sqrt{n}} $$

```python
# alpha = 0.95, then z = 1.96 and n=50

z = 1.96
n = len(bill_length_adelie)

lower_limit_eq = sample_mean - ((z * sample_std) / np.sqrt(n))
upper_limit_eq = sample_mean + ((z * sample_std) / np.sqrt(n))

print(f'The Confidence Interval of the average bill_length of Adelie is between {lower_limit_eq} and {upper_limit_eq}')
```
